#TO find probability using scipy


```python
import scipy
from scipy import stats
import pandas as pd
import numpy as np
```

```python
#calculating the probability that a student chosen at random will have less than 70 marks.
stats.norm.cdf(70,loc=60,scale=10)
```

```python
#probabilty that a student has more than 80 marks
s=stats.norm.cdf(80,loc=60,scale=10)
ans=1-s #this formula always gives the lessthan value.To find greater than use 1-(the answer)
print(ans)
```
