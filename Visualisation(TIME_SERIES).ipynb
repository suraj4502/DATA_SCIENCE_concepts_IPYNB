#### Line plot

```python
# create a line plot
from pandas import read_csv
from matplotlib import pyplot
```

```python
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
series.head()
```

```python
series.plot(figsize=(18,8),c = '#00C957')
pyplot.ylabel("Temperature")
pyplot.title("Time series")
pyplot.show()
```

#### Histogram and Density Plots

```python
#density plot
series.plot(kind='kde',figsize=(12,8),c='#00C957')
pyplot.show()
```

#### Box and Whisker Plots by Interval

```python
# create a boxplot of yearly data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

n = 365
data = pd.read_csv('/content/daily-minimum-temperatures.csv',index_col='Date', parse_dates=True)
data
```

```python
# Year and Month
data['Date'] = data.index
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,
                       figsize=(20, 8))
sns.boxplot(data['Year'],
            data["Temp"], ax=ax[0])
ax[0].set_title('Year-wise Box Plot for A',
                fontsize=20, loc='center')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Temperature')


sns.boxplot(data['Month'],
            data["Temp"], ax=ax[1])
ax[1].set_title('Month-wise Box Plot for A',
                fontsize=20, loc='center')
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Temperature');
```

#### Lag plot

```python
# create a scatter plot
from pandas import read_csv
from matplotlib import pyplot
from pandas.plotting import lag_plot
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
lag_plot(series)
pyplot.show()
```

```python
# create an autocorrelation plot
from pandas import read_csv
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_acf
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
plot_acf(series,lags=30)
pyplot.show()
```
